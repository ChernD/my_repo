```python
coefs = []
for col in features.columns:
    print(col)
    coef = np.corrcoef(features['{}'.format(col)], features.radiant_win)[0, 1]
    print(coef)
    coefs.append([coef])
```

```python
porog = 0.1

coefs = np.array(coefs)
print('corr>{}:'.format(porog))
print(np.where(coefs>porog))
print(coefs[coefs>porog])
print('corr<-{}:'.format(porog))
print(np.where(coefs<-porog))
print(coefs[coefs<-porog])
```

```python
X = features.drop(columns=['duration', 'radiant_win', 'tower_status_radiant', 'tower_status_dire',
                           'barracks_status_radiant', 'barracks_status_dire'])
y = features.radiant_win
```

```python
plt.stem(X.isnull().sum()[X.isnull().sum()>0])
print(X.isnull().sum()[X.isnull().sum()>0])
```

Итого строка "first_blood_player2" имеет наибольшее количество пропусков. 

```python
X.fillna(10000, inplace=True)
```

```python
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import KFold
from sklearn.metrics import roc_auc_score
import sklearn
```

```python
sorted(sklearn.metrics.SCORERS.keys())
```

```python
estimators_arr = [10, 20, 30]
```

```python
divider = KFold(5, shuffle=True, random_state=1)
for est in estimators_arr:
    gb = GradientBoostingClassifier(n_estimators=est)

    score = sum(cross_val_score(gb, X, y, cv=divider.split(X,y), scoring='roc_auc', n_jobs = -1))/5
    print(score)
```

n_estimators=100, который по умолчанию, ведет себя лучше

```python
gb = GradientBoostingClassifier(n_estimators=100)

score = sum(cross_val_score(gb, X, y, cv=divider.split(X,y), scoring='roc_auc', n_jobs = -1))/5
print(score)
```

```python
gb.fit(X, y)

test_data = pd.read_csv('final data/features_test.csv', index_col='match_id')

test_data.fillna(10000, inplace=True)

predicted_test_values = gb.predict(test_data)
```

```python
print(predicted_test_values)
pred1 = ['radiant' if x==1 else 'dire' for x in predicted_test_values]
sns.countplot(pred1)
```
