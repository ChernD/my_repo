```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import scipy.signal as signal
import scipy.fftpack as fp
from math import pi
```

```python
f1=40; A1=0.9;
f2=60; A2=0.5;
f3=70; A3=1.5;

Fs=250; 
Ts=2; 

t=np.linspace(0,Ts,Ts*Fs)
s1 = A1*np.sin(2*pi*f1*t)
s2 = A2*np.sin(2*pi*f2*t)
s3 = A3*np.sin(2*pi*f3*t)
sn = s1 + s2+ s3
sp = s2 + s3
Sf=fp.fft(sn);
```

```python
ft=35
fk=55
WDp = 2 * ft / Fs
WDs = 2 * fk / Fs
print(WDp, WDs)
d1=0.03
d2=0.01
rp= -20 * np.log10(1-d1)
rs= -20 * np.log10(d2)
```

```python
R, WDn = signal.buttord(WDp,WDs,rp,rs) 
print(WDn)
b, a = signal.butter(R, WDn, btype='low')
print('R = ', R)
```

```python
z1 = signal.lfilter(b, a, s1)
z2 = signal.lfilter(b, a, s2)
z3 = signal.lfilter(b, a, s3)
zp = signal.lfilter(b, a, sp)
zn = signal.lfilter(b, a, sn)
```

```python
def plotter(s_sign, z_sign, time):
    plt.plot(time, s_sign)
    plt.plot(time, z_sign)
    plt.grid()
    plt.ylim((-3, 3))
    plt.legend(('input signal', 'output signal'))
```

```python
plotter(s1, z1, t)
```

```python
plotter(s2, z2, t)
```

```python
plotter(s3, z3, t)
```

```python
plotter(sp, zp, t)
```

```python
plotter(sn, zn, t)
```
