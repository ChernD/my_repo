## Цифровые БИХ-фильтры на python

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import scipy.signal as signal
```

```python
Fs = 16000
ft = 4000
fk = 3000
WDp = 2 * ft / Fs
WDs = 2 * fk / Fs
print(WDp, WDs)
rp = 0.2
rs = 60
d2 = 10**(rs/-20)
d1 = 1 - 10**(rp/-20)
f = np.linspace(0, Fs/2, 1000)
print(rp, rs)
```

#### Фильтр Баттерворда

```python
R1, WDn = signal.buttord(WDp,WDs,rp,rs) 
b, a = signal.butter(R1, WDn, btype='high') 
Ha1 = signal.freqz(b,a,worN=f,fs=Fs)
Mag1 = np.abs(Ha1)
print("для фильтра Баттерворта R = ", R1)
plt.plot(*Mag1)
plt.ylabel('Amplitude')
plt.xlabel('Frequency [rad/sample]')
plt.grid()
```

#### Фильтр чебышева 1 рода

```python
R2, WDn = signal.cheb1ord(WDp,WDs,rp,rs) 
b, a = signal.cheby1(R2, rp, WDn, btype='high') 
Ha1 = signal.freqz(b,a,worN=f,fs=Fs)
Mag1 = np.abs(Ha1)
print("для фильтра Чебышева R = ", R2)
plt.plot(*Mag1)
plt.ylabel('Amplitude')
plt.xlabel('Frequency [rad/sample]')
plt.grid()
```

#### Фильтр чебышева 2 рода

```python
R3, WDn = signal.cheb2ord(WDp,WDs,rp,rs) 
b, a = signal.cheby2(R2, rp, WDn, btype='high') 
Ha1 = signal.freqz(b,a,worN=f,fs=Fs)
Mag1 = np.abs(Ha1)
print("для фильтра Чебышева R = ", R3)
plt.plot(*Mag1)
plt.ylabel('Amplitude')
plt.xlabel('Frequency [rad/sample]')
plt.grid()
```

#### Фильтр Золотарева-Кауэра

```python
R4, WDn = signal.ellipord(WDp,WDs,rp,rs) 
b, a = signal.ellip(R4, rp, rs, WDn, btype='high') 
Ha1 = signal.freqz(b,a,worN=f,fs=Fs)
Mag1 = np.abs(Ha1)
print("для фильтра Золотарева-Кауэра R = ", R4)
plt.plot(*Mag1)
plt.ylabel('Amplitude')
plt.xlabel('Frequency [rad/sample]')
plt.grid()
```
