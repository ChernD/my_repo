###  Синтез цифровых БИХ-фильтров с использованием MATLAB (Octave)


Вариант 2:

Fs=16 кГц;

ft=4 кГц;

fk=3 кГц;

а0=60 дБ;

Δа = 0,2 дБ.

```octave
pkg load signal
```

```octave
Fs=16000;         
ft=4000; 
fk=3000; 
WDp=ft/(Fs/2) 
WDs=fk/(Fs/2) 
rp= 0.2;
rs= 60;
d2 = 10^(rs/-20); % 0.0031622776601683794
d1 =1 -  10^(rp/-20); % 0.02276277904418933 
f=0:((Fs/2)/1000):Fs/2;
```

```octave
[R1,WDn1]=buttord(WDp,WDs,rp,rs);
[b1,a1]=butter(R1,WDn1, 'high');
Ha1=freqz(b1,a1,f,Fs); 
Mag1=abs(Ha1);
plot(f,Mag1);
grid on
title ('Butter')
```

```octave
[R2,WDn2]=cheb1ord(WDp,WDs,rp,rs);
[b2,a2]=cheby1(R2,rp,WDn2,'high');
Ha2=freqz(b2,a2,f,Fs); 
Mag2=abs(Ha2);
plot(f,Mag2);
grid on
title ('Cheb1 1')
```

```octave
[R3,WDn3]=cheb2ord(WDp,WDs,rp,rs);
[b3,a3]=cheby2(R3,rs,WDn3,'high');
Ha3=freqz(b3,a3,f,Fs);
Mag3=abs(Ha3);
plot(f,Mag3);
grid on
title ('Cheb 2')
```

```octave
[R4,WDn4]=ellipord(WDp,WDs,rp,rs);
[b4,a4]=ellip(R4,rp,rs,WDn4,'high');
Ha4=freqz(b4,a4,f,Fs);
Mag4=abs(Ha4);
plot(f,Mag4);
grid on
title ('ellip')
```
